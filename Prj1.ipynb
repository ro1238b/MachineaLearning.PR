```python
# Convert 'X' and 'Y' columns to numeric type after replacing commas and spaces
popu['X'] = popu['X'].astype(str).str.replace(',', '').str.replace(' ', '').astype(float)
popu['Y'] = popu['Y'].astype(str).str.replace(',', '').str.replace(' ', '').astype(float)

X = popu['X'].values
Y = popu['Y'].values

plt.scatter(X, Y)
plt.xlabel("year")
plt.ylabel("Population")
plt.title("Scatter Plot of Year vs Population")
plt.show()

```

```python
X = popu['X'].values.reshape(-1, 1)
Y = popu['Y'].values
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.5,random_state=100)
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
Y_pred=lr.predict(X_test)
Y_pred
```
